To assess the impact of Airbnb on the housing market, it is important to consider both the number of short-term rental listings and the availability of housing for long-term rental.
One way to capture this availability is through the vacancy rate. The vacancy rate measures the proportion of housing that is currently unoccupied. The vacancy rate is commonly used to reflect housing resource utilization. A high vacancy rate may indicate housing oversupply or a market with excess supply relative to demand, while a low vacancy rate may suggest that housing is in short supply and the rental market is tight.
The Airbnb ratio is calculated as the number of Airbnb listings divided by total dwellings, which in this case used to reflect the scale of short-term rental activity in the overall housing market.
Combining Airbnb ratio and vacancy rate helps assess the relative intensity of market impact: Areas with high Airbnb ratio and low vacancy rate have a relatively large short-term rental presence compared with limited potential long-term housing, whereas in areas with high Airbnb ratio and high vacancy rate indicating that these areas may be able to accommodate short-term rentals without putting additional pressure on the long-term housing supply.
Taken together, these indicators highlight the heterogeneity among boroughs and the varying degree to which short-term rentals may impact local housing availability.

# limitations

Vacancy data from the rental market is not available, so the vacancy rate is calculated based on all dwellings, not just those available for rental. Some vacant homes may be owner-occupied or otherwise unavailable for renting, so the measure is an approximation, so as Airbnb ratio. Other factors affecting housing availability or Airbnb distribution (e.g., local regulations, rental demand, property prices) are not included in this part.
The dwelling stock data are from 2021, which could cause slight deviations. However, changes in total housing stock over time are relatively modest, so these changes are expected to have minimal impact on the findings.
